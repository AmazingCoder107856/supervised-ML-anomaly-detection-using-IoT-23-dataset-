```python
# Import necessary libraries and modules
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
import time
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
import warnings
```

```python
df = pd.read_csv("iot23_combined_new.csv")
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
df
```

```python
df['label'].value_counts()
```

```python
warnings.filterwarnings('ignore')
```

```python
X = df[['duration', 'orig_bytes', 'resp_bytes', 'missed_bytes', 'orig_pkts', 'orig_ip_bytes', 'resp_pkts', 'resp_ip_bytes']]
Y = df['label']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=10, test_size=0.2)
```

```python
clf=RandomForestClassifier(n_estimators =100, random_state=10)
```

```python
start = time.time()
print('program start...')
print()

clf.fit(X_train,Y_train)
print()

print('prediction:')
y_pred=clf.predict(X_test)
print(y_pred)
print()

print('Score:')
score = clf.score(X_test,Y_test)
print(score)

end = time.time()
print('program end...')
print()
print('time cost: ')
print(end - start, 'seconds')
```

```python
print("Classifiction Report :")
print(classification_report(Y_test, y_pred))
```

```python
# using GridSearchCV from sklearn.model_selection to perform hyperparameter 
# tuning on a Random Forest classifier. 
# This is a good approach to find the optimal parameters for your model.
param_grid = { 
    'max_features': ['auto', 'sqrt', 'log2',0.2],
    'max_depth' : [5,6,7,8],
    'criterion' :['gini', 'entropy']
}
```

```python
# This line is initializing the GridSearchCV object with your classifier (clf), 
# the parameter grid (param_grid) that you want to search, and the number of folds (cv=5) for cross-validation.
CV_rfc = GridSearchCV(estimator=clf, param_grid=param_grid, cv= 5)

# This line is fitting the model to your training data (X_train, Y_train). 
# It will run the cross-validation for each combination of parameters in your grid.
CV_rfc.fit(X_train, Y_train)
```

```python
# After fitting, you can get the best parameters found by GridSearchCV 
CV_rfc.best_params_
```

```python
# After fitting, you can get the best estimator found by GridSearchCV 
CV_rfc.best_estimator_
```
